```python
import json
from collections import Counter
import spacy
import pandas as pd

# Load the spaCy language model
nlp = spacy.load("en_core_web_sm")
```

```python
# Read and parse the JSON file
with open('tweets.json', 'r') as json_file:
    data = json.load(json_file)
```

```python
# Initialize a counter for entity frequency
entity_counter = Counter()
```

```python
# Process each tweet
for tweet_data in data:
    tweet = data[tweet_data]['tweet_text']
    doc = nlp(tweet)
    entities = [ent.text for ent in doc.ents]
    entity_counter.update(entities)
```

```python
entity_counter
```

```python

# Sort entities by frequency in descending order
sorted_entities = [(entity, freq) for entity, freq in entity_counter.most_common()]

# Create a DataFrame for the results
df = pd.DataFrame(sorted_entities, columns=['entity', 'frequency'])

# Save the results to a CSV file
df.to_csv('objective1.csv', index=False)
```

```python

import pandas as pd
from transformers import pipeline

# Load a pre-trained sentiment analysis model from Hugging Face Transformers
sentiment_analyzer = pipeline("sentiment-analysis")

# Create a dictionary to store sentiment scores
sentiment_scores = {}

# Process each tweet
i = 0

for tweet_data in data:
    i = i+1
    if i<50:
        author = data[tweet_data]['tweet_author']
        tweet = data[tweet_data]['tweet_text']
        sentiment = sentiment_analyzer(tweet)  # Analyze sentiment

        # Store the sentiment score in the dictionary
        sentiment_scores[author] = sentiment_scores.get(author, [])
        sentiment_scores[author].append(sentiment[0]['label'])

```

```python
sentiment_scores
```

```python
sentiment_author_list = []
for author, sentiments in sentiment_scores.items():
    for sentiment in sentiments:
        sentiment_author_list.append([author, sentiment])
```

```python
# Convert the sentiment scores dictionary to a DataFrame
df = pd.DataFrame(sentiment_author_list)
df.rename(columns = {0:'Entity',1:'Sentiment'}, inplace = True)
# Save the results to a CSV file
# 
```

```python
df.to_csv('sentiment_scores.csv')
```
